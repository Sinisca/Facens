# Exercício 1 - Ligações para o 911

### Nos EUA, o número 911 é destinado à emergências. O [Kaggle](https://www.kaggle.com/mchirico/montcoalert) disponibilizou dados de ligações para o serviço:

* lat: Variável String, Latitude
* lng: Variável String, Longitude
* desc: Variável String, Descrição da Chamada de Emergência
* zip: Variável String, CEP
* título: Variável String, Título
* timeStamp: Variável String, AAAA-MM-DD HH: MM: SS
* twp: Variável String, Township
* addr: Variável String, Endereço
* e: Variável String, variável Dummy (sempre 1)



```python
import pandas as pd
```

### Leia o arquivo csv (911.csv) como um dataframe chamado _calls_

```python
calls = pd.read_csv("911.csv")
calls
```

```python
calls.info()
```

### Quais são os top 5 CEPs nas chamadas 911?

```python
calls.groupby('zip').agg({'e': 'count'}).sort_values(by = 'e', ascending=False).head(5)
```

### Quais são os 5 principais municípios nas chamadas 911?

```python
#primeira tentativa
calls_top5 = calls.groupby('twp').agg({'e': 'count'}).sort_values(by = 'e', ascending=False).head(5)
print (calls_top5)


#segunda maneira
calls_top5 = calls['twp'].value_counts().head(5)
print('\n',calls_top5)
```

### Quantos códigos de título únicos existem?

### Na coluna _title_ existem Razões / Departamentos especificados antes do código do título. Estes são _"EMS"_, _"Fire"_ e _"Traffic"_. Crie uma nova coluna chamada _Reason_ que conterá esse valor em cada linha

#### Por exemplo, se o valor da coluna do título for EMS: BACK PAINS / BLESSOR, o valor da coluna Reason será EMS.

```python
calls["Reason"] = calls["title"].str.split(":").str[0]
calls
```

```python
separator = lambda x : x.split(":")[0]
calls['Reason'] = calls.title.apply(separator)
calls.head()
```

```python
calls['Reason'] = calls['title'].str.extract('([A-Z]\w{0,})', expand=True)
calls
```

### Qual é a razão mais comum para uma chamada do 911 com base nessa nova coluna?

```python
calls["Reason"].value_counts().head(1)
```

### Qual é o tipo de dados dos objetos na coluna _timeStamp_?

```python
print (calls['timeStamp'].dtype)
```

### Converta a coluna _timeStamp_ para DateTime. Use [pd.to_datetime](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) para essa operação.

```python
calls['timeStamp'] = pd.to_datetime(calls['timeStamp'])
calls
```

### Crie 3 novas colunas chamadas _Hour_, _Month_ e _Day of Week_. Você criará essas colunas com base na coluna timeStamp

```python
calls['Hour'] = pd.DatetimeIndex(calls['timeStamp']).hour
calls['Month'] = pd.DatetimeIndex(calls['timeStamp']).month
calls['Day of Week'] = pd.DatetimeIndex(calls['timeStamp']).weekday
calls
```

```python
calls['Hour'] = calls.timeStamp.apply(lambda x: x.hour)
calls['Month'] = calls.timeStamp.apply(lambda x: x.month)

#week day where monday is 0 and sunday is 6
calls['Day of Week'] = calls.timeStamp.apply(lambda x: x.dayofweek)

calls.head()
```

### Observe como o dia da semana é um número inteiro de 0-6. Mapeie os nomes das seqüências reais para o dia da semana, nessa mesma coluna:

- 0 => Mon
- 1 => Tue
- 2 => Wed
- 3 => Thu
- 4 => Fri
- 5 => Sat
- 6 => Sun

```python
calls['Day of Week'] = calls['Day of Week'].map({0:"Mon", 1:"Tue", 2:"Wed", 3:"Thu", 4:"Fri", 5:"Sat", 6:"Sun"})
calls
```

```python
calls['Day of Week'] = calls['timeStamp'].dt.day_name().str[0:3]
calls
```

### Qual a quantidade de chamadas de cada motivo (_Reason_) por mês?

```python
by_month = calls.groupby(by=['Reason', 'Month']).sum()
by_month['e']
```

```python
calls.groupby('Month').agg({'Reason':'value_counts'})
```

### Qual foi o dia com a maior quantidade de chamadas? Quantas chamadas foram executadas neste dia?

```python
by_day = calls.groupby(by=[calls.timeStamp.dt.date]).sum()
by_day['e'].sort_values(ascending=False).head(1)
```

```python
calls['Date'] = calls['timeStamp'].apply(lambda x: x.strftime("%d/%m/%y"))

#calls.head()

calls['Date'].value_counts().sort_values(ascending=False).head(1)
```

### Levando-se em consideração apenas atendimentos feitos às sextas-feiras, qual a hora do dia com o maior número de chamadas?

```python
calls_friday = calls.groupby(by=["Day of Week","Hour"]).sum()
calls_friday = calls_friday['e'].sort_values(ascending=False)
calls_friday.loc['Fri'].head(1)
```

```python
calls[calls['Day of Week']=='Fri']['Hour'].value_counts().head(1).index[0]
```

### Levando-se em consideração os atendimentos relacionados a incêndio (_Reason_ = _Fire_) qual a quantidade de chamados aos sábados?

```python
calls_sat = calls.groupby(by=['Reason','Day of Week']).sum()
calls_sat = calls_sat['e']
calls_sat.loc['Fire','Sat']
```

```python
calls[calls.Reason =='Fire']['Day of Week'].value_counts().loc['Sat']
```
